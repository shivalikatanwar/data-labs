#### Import all the libraries that are necessary.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
plt.subplot(1,2,1)
plt.plot(x,y,color='g',ls='--')
plt.subplot(1,2,2)
plt.plot(x,z,color='y')
```

```python
fig, [ax1,ax2]=plt.subplots(1,2)
ax1.plot (x,y,color='g',lw=4,ls='--')
ax2.plot (x,z,color='y',lw=2)
plt.show()
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
fig, [ax1,ax2]=plt.subplots(1,2,figsize=(12,4))
ax1.plot (x,y,color='g',lw=4,ls='--')
ax2.plot (x,z,color='y',lw=2)
plt.show()
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
fig, [ax1,ax2]=plt.subplots(1,2)
ax1.plot (x,y,color='g',lw=4,ls='--')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')
ax2.plot (x,z,color='y',lw=2)
ax2.set_xlabel('x label')
plt.show()
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
fig, [ax1,ax2]=plt.subplots(1,2)
ax1.plot (x,y,x,np.exp(x),color='g',lw=4,ls='--')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')
ax1.set_title('Normal')

ax2.plot (x,y,x,np.exp(x),color='y',lw=2)
ax2.set_xlabel('x label')
ax2.set_yscale('log')
ax2.set_title('Logarithmic scale')
plt.show()
```
